```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./movies.csv")
df.dropna(inplace = True)
```

```python
#count of the number of movies in each mpaa_rating
ax = sns.countplot(x = df.mpaa_rating)
ax.set_xticklabels(labels = df.mpaa_rating.unique(), rotation = "vertical")
plt.show()
```

```python
#how many movies are in each genre. 
sns.countplot(x = df.genre)
plt.show()
```

```python
#Production budget count with histoplot
sns.histplot(x = df.production_budget, bins = 50, color = "red")
plt.title("production budget count")
plt.show()
```

```python
#Worldwide gross with histoplot
sns.histplot(x = df.worldwide_gross)
plt.show()
```

```python
#scatter plot with the line of best fit
sns.regplot(data = df, x = "domestic_gross", y = "worldwide_gross")
plt.show()
```

```python
#boxplot 
sns.boxplot(data = df, x = "mpaa_rating", y = "worldwide_gross")
plt.show()
```

```python
#facet grid
ax = sns.FacetGrid(data = df,col ="genre")
ax.map(sns.countplot,x = df.mpaa_rating, order = None)
plt.show()
```

```python
#juxtaposed countplot
sns.countplot(x = df.genre, hue = df.mpaa_rating, dodge = True)
plt.savefig(fname= "figure-juxtaposed")
plt.show()
```

```python
ax2 = sns.FacetGrid(data = df, col= "genre")
ax2.map(sns.histplot, x = df.production_budget,color = "orange")
plt.show()
```
